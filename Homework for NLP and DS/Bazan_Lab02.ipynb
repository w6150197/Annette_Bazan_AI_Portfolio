### Data Structures in Python:

1. Lists: They are ordered, accept duplicates, mutable
2. Tuples: They are ordered, accept duplicates, immutable
3. Sets: They are unordered, don't accept duplicates, immutable:(except for add and remove elements)
4. Dictionary: They are unordered, don't accept duplicates, mutable, and the elements of a dictionary are in pairs of key/value.

**What is ordered? An ordered element have a fixed position; therefore, we can index(accessing a single element) an element in an ordered data structure.**

**What is duplicates? When the same values have different positions.**

**What is mutable? To be able to modify or change the value of an element in a data structure.**

### 1. Lists

```python
L = [1, 2, 'Annette', 3.14, True, 2]
L
```

```python
type(L)
```

#### Indexing and Slicing
1. Indexing: Accessing a single element
2. Slicing: Accessing a range of elements

```python
# Let's index L and display 'Annette'
L[2]
```

```python
L[-2]
```

```python
# Slicing the range of 2,'Annette', 3.14 in L
L[1:4]
```

```python
# to access all the elements
L[::]
```

```python
# Reverse the order of the elements
L[::-1]
```

```python
# adding an element: append() method adds an element at the end of a list
L.append(50)
L
```

```python
# deleting elements from a list
# remove() method deletes elements by their values not positions
L.remove(2)
L
```

```python
L2 = L.copy()
L2
```

```python
# making an empty list
L2.clear()
L2
```

```python
L[3] = False
L
```

### 4. Dictionaries

```python
D = {1:'a', 'B':2, 5:3.14, 'school':'HCC', 'Annette':True}
D
```

```python
# we can acess a value by its key
D['school']
```

```python
D[5] = 'float'
D
```

```python
# AttributeError: 'dict' object has no attribute 'add'
# D.add('key:value')
```

```python
# adding one pair to D
D['newKey'] = 'newValue'
D
```

```python
# adding more than one pair to D using the update() method
D.update({'cat':9, 'summer':'hot', 'TX':'Austin'})
D
```

```python
# deleting a pair in D
D.pop('cat')
D
```

```python
# another way to remove a pair in dictionary is using del keyword
del D['B']
D
```

```python
# Shows only the keys
D.keys()
```

```python
# Shows only the values
D.values()
```

```python
# TypeError: unhashable type: 'list'
# We can't assign a mutable object as a key
# immutable_key = {1: 'a', 5: 'float', [1,2,3,4,5,]: 'HCC',}
```

```python
immutable_key_tuple = {1:'a', 5: 'float', (1,2,3,4,5): 'HCC'}
immutable_key_tuple
```

**Observation:**
1. We cannot assign a list as a key since a list is mutable.
2. Whereas, we can use a tuple as a key in a dictionary since a tuple is immutable.
