### Python data structures are:

1. Lists - They are:
* Ordered,
* Can have duplicates,
* Mutable (items are changeable),
* Example, [1, 2, 3, a, 2]

2. Tuples - They are:
* Ordered,
* Can have duplicates,
* Immutable (items are unchangeable),
* Example, (2, 6, 8, 6)

3. Sets - They are:
* Unordered,
* No duplicates (unique values only),
* Immutable (addable and removeable only),
* Example, {9, 1, 0, 5}

4. Dictionaries - They are:
* Unordered,
* No duplicates (unique key/value pairs)
* Mutalbe (items are changeable)
* Example, {4:'a', 'dog':1, 'N':'Annette'}

-- **What is ordered? Ordered collection of data means the elements are in order. They have a specific position when we index each element.**

-- **What is mutable? Mutable means the items are changeable, modifiable.**

-- **What are duplicates? Duplicates elements means we can have the same elements in different positions.**



```python
# this is a list
L = [1, 3, 4.9, 'Annette', 3]

# this is a tuple
T = (1, 3, 4.9, 'Annette', 3)

# this is a set
S = {1, 3, 4.9, 'Annette', 3}

# this is a dictionary
D = {14: 'one and 4', 'N':'Annette', 'B':19}
```
