### Exploratory Data Analysis (EDA)

##### What is EDA?
1. EDA is a process in data analysis used to analyze and investigate datasets and find out about the main characteristics.
2. Visualization tools can offer a graphic picture of the EDA results.
3. EDA helps data analysts to discover patterns and check assumputions.
4. EDA is primarily used to better understand the dataset variables and the relaitionships between them.
5. It was developed by American mathematician John Tukey in the 1970s.
6. EDA also helps the management team in answering their questions.
7. When EDA is done, it can be used for more sophisticated data analysis such as ML modeling.

##### Types of EDA:
1. Univariate
2. Bivariate
3. Multivariate

##### EDA Tools:
1. Graphical(data visualization): Plots, Charts, Heatmaps
2. Nongraphical: Examining the realtionship between the variables.

##### EDA most popular languages:
* Python
* R

### Steps in a data analysis project:
1. Data background(context)
