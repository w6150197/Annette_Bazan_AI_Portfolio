**What is EDA (Exploratory Data Analysis)?**
1. EDA is a process in data analytics use to analyze and investigate datasets and find out the main characteristics.
2. EDA helps data analysts to discover patterns in the dataset and examine their assumptions.
3. EDA is also used to find out about the relationship between the datasets variables.
4. It was developed by the American mathametician, John Turkey in the 1970s.
5. EDA helps analysts to answer the management team questions.
6. When EDA is done,  the dataset can be used for more sophisticated applications of AI such as ML.

**Types of EDA**
1. Univariate: Examining one variables
2. Bivariate: Examining two variables
3. Multivariate: Examining more than two variables

**EDA Tools:**
1. Graphical EDA (data visualization): plots, graphs, charts
2. Non-Graphical EDA: Examining the relationship between the dataset variables.

**EDA Most Popular Languages:**
1. Python
2. R

### Data Indexers
- There are 2 types of indexers:
 1. iloc indexer -loc=location: This is implicit indexing (not clearly described/stated). It is integer-based and refers to the position of the element. It starts with 0.
 2. loc indexer: This is explicit indexing (clearly stated/described). It is the user-defined name and acts as a locator of the element.

```python
# 2. What is the movie sitting in position 300?
movies.iloc[300]
```

```python
# 3. What is the information about the movie"Forrest Gump"?
movies.loc['Forrest Gump']
```

```python
# 4. Is there only one version of the movie "101 Dalmatians"?
movies.loc["101 Dalmatians"]
```

```python
# 5. What are the movies created by Universal Studio?
# Let's do double indexing
universal = movies[movies['Studio']=='Universal'] 
universal
```

```python
# 6. What are the top 5 movies by Universal Studio?
universal.head()
```
