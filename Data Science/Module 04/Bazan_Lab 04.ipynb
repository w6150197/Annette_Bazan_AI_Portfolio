### Arrays Broadcasting
1. Broadcasting performs math operations more effectively and faster on **arrays with different shapes**
2. Broadcasting also uses the "vectorization" concept.

```python
# 1. Broadcastin on 1-D array using a scaler(number)
h = np.array([4,6,9])
h
```

```python
# The scaler is 5 and it adds it to the array in h
h1 = h + 5
h1
```

```python
# The scaler is 2 and it is dedcuting from every element in h
h2 = h - 2
h2
```

```python
# Every element in h is multiplyed by 3(scaler) 
h3 = h * 3
h3
```

```python
# Every element in h is divided by 2(scaler) simple division
h4 = h / 2
h4
```

```python
# floor division, drops the remainder 
h5 = h // 2
h5
```

```python
# 2. Broadcasting on arrays with different shapes
h6 = np.array ([[1],[2],[3]])
h6
```

### NumPy Universal Functions (ufuncs)
1. Univerasal funcitons (ufuncs) are highly effective in performing math operations on arrays, by converting iterative objects into vectors.
2. Ufuncs also use the vectorization concept.

```python
# Let's create x and y arrays
x = np.array([5,8,3,9])
x
```

```python
y = np.array([2,6,7,4])
y
```

```python
# Let's check out if a fucntion is ufunc or not
type(np.sum)
```

```python
type(np.add)
```

```python
# 1. ufunc: addition
ufunc1 = np.add(x,y)
ufunc1
```

```python
# 2. ufunc: subtraction
ufunc2 = np.subtract(y,x)
ufunc2
```

```python
# 2. ufunc: subtraction (order matters)
ufunc2_1 = np.subtract(x,y)
ufunc2_1
```

```python
x
```

```python
y
```

```python
# 3. ufunc: multiplication 
ufunc3 = np.multiply(y,x)
ufunc3
```

```python
# 4. ufunc: division
ufunc4 = np.divide(x,y)
ufunc4
```

```python
# 4. ufunc: divmod() 
ufunc4_1 = np.divmod(x,y)
ufunc4_1
```

```python
# 5. ufunc: power() exponents
ufunc5 = np.power(x,y)
ufunc5
```

```python
# 6. ufunc: remainder()  
ufunc6 = np.remainder(x,y)
ufunc6
```

```python
# 6. ufunc: mod()  
ufunc6_1 = np.mod(x,y)
ufunc6_1
```

```python
# 7. ufunc:absolute()
x1= np.array([8,-3,2,-5,-6])
ufunc7 =np.absolute(x1)
ufunc7
```

```python
# 8. ufunc: truncation:removing decimals : trunc()
y1 = np.array([-3.1666,2.457])
ufunc8 =np.trunc(y1)
ufunc8
```

```python
# 8. ufunc: fixed() 
ufunc8_1 = np.fix(y1)
ufunc8_1
```

```python
# 9. ufunc: around(): rounding decimals
ufunc9 = np.around(6.1666)
ufunc9
```

```python
# 10.ufunc: floor(): rounding decimals( don't forget to enclose in square brackets for more than one value)
# rounding decimals to the nearest lower integer
ufunc10 = np.floor([6.1666,-2.444])
ufunc10
```

```python
# 11. ufunc: log2(): getting log based 2
x2 = np.array ([5,10,15])
ufunc11 = np.log2(x2)
ufunc11
```

```python
# 11. ufunc: log10(): getting log based 10
ufunc11_1 = np.log10(x2)
ufunc11_1
```

```python
# 12. ufunc: prod(): getting the product
ufunc12 = np.prod(x2)
ufunc12
```

```python
# 12. ufunc: prod(): getting the product
x2 = np.array([5,10,15])
y2 = np.array([3,7,2])
ufunc12_1 = np.prod([x2,y2])
ufunc12_1
```

```python
# 13. ufunc: lcm(): lowest common multiple 
# reduce(): displays a single element 
x2 = np.array([5,10,15,18])
ufunc13 = np.lcm.reduce(x2)
ufunc13
```

```python
# 14. ufunc: gcd(): greatest common divisor 
x2 = np.array([20,10,32,16])
ufunc14 = np.gcd.reduce(x2)
ufunc14
```
