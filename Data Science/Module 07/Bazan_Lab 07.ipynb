### Introduction:

**What is data visualization (VIZ)?**
- Graphical presentation of data.

**What are the benefits of data visualization (VIZ)?**
1. Helps with discovering the data patterns
2. Helps with more effective communication
3. Helps non-technical people to understand the data better or easier
4. Helps with understanding the data distribution
5. Helps with discovering the current trends in the dataset

**What are the types of data visualization?**
There are 3 types of data visulaization.
1. Univariate Visualization: plotting one variable.
2. Bivariate Visualization: plotting two variables.
3. Multivariate Visualization: plotting more than two variables.

**What are popular types of data visualization plots (Viz) tools/techniques?**
1. Line Plots
2. Bar Plots / Box chart
3. Pie Charts
4. Scatter Plots
5. Histograms
6. Heatmaps

**What are the popular libraries used for data visualization(Viz)?**
1. matplotlib.pyplot
2. seaborn


```python
# Importing libraries

# general libraries
import numpy as np
import pandas as pd
import math

# data visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# to suppress warnings library
import warnings
warnings.filterwarnings('ignore')

# 1. These Python libraries for data visualizations are said to be platform independence:
# This means that we can use these libraries on any OS(operating system): Windows, Linux, and IOS
# 2. To enhance the performance of matplotlib, an API has been added to it. 
# This API is pyplot.
# * API: Application Programming Interface: Acts as a bridge that connects two or more applications/programs.
```

```python
# 1. Line Plots: These are the simplest plots
# They display the progress of an event over time

# Creating the dataset
a = np.linspace(0, 10)

# Plotting the dataset: The sin and cos of the numbers 0-10
plt.plot(a, np.sin(a)) # The sin of 0 is 0
plt.plot(a, np.cos(a)); # The cos of 0 is 1
```

```python
# Creating the dataset
x = np.array([0, 5])
y = np.array([0, 20])

# Plotting the dataset:
plt.plot(x, y, color='red');
```

```python
# Creating the dataset
y = np.array([12, 0, 9, 2, 14])

# Plotting the dataset: The sin and cos of the numbers 0-10
plt.plot(y,color ='green', linestyle = 'dashed', marker = '*');
```

```python
# Creating the dataset
x = pd.Series(['Sunday', 'Monday', 'Tuesday', 'Thursday'])
y = pd.Series(['Running','Tennis', 'Football', 'Biking'])

# Plotting the dataset: The sin and cos of the numbers 0-10
plt.plot(x,y, color = 'orange', marker = 'o', linestyle = ':')
plt.title('Days Vs Sports')
plt.xlabel('Days')
plt.ylabel('Sports')
plt.grid();
```

```python
# 2. Bar Plots: These are good to compare various categories.

# Creating the dataset
y = np.array([5, 8, 10, 2])
x = np.array(['Five', 'Eight', 'Ten', 'Two'])

# Plotting the dataset
plt.bar(x,y, color = 'r')
plt.title('Numbers Vs Written Forms');
```

```python
# 3. Pie Charts: These are good to show the proportions over the whole

# Creating the dataset
y = np.array([5, 20, 10, 8, 25])
cars = ['Benz', 'Ford', 'Honda','Tesla', 'Maserati'] # Labels 
spaces = [0, 0.2, 0, 0, 0.1] #distances
kolrs = ['b', 'c', 'm','y', 'g']

# Plotting the dataset: The sin and cos of the numbers 0-10
plt.pie(y, labels = cars, explode = spaces, colors = kolrs )
plt.legend();
```

**Color Codes**
1. r for red
2. g for green
3. b for blue
4. c for cyan
5. y for yellow
6. m for magenta
7. k for black
8. w for white

**Heatmaps:**

1. Heatmaps are seaborn library plot that displays correlation
2. What is correlation? Correlation is the relationship between variables
3. Correlation can be positive or negative
4. If the variables move together in either directions (up or down), we call this a positive correlation.
5. If the variables move in opposite directions(one goes up and the other goes down), we call this a negative correlation.
6. A heatmap is designed for numeric values only.
7. The range of correlation in the heatmap goes from -1 to +1.

```python
# Let's create a new dataset and name it num_df that contains only numeric variables of Titanic
num_df = titanic[['Age','Fare','Pclass', 'SibSp', 'Parch','Survived']]
num_df
```

```python
# Correlation table 
num_df.corr()
```

**Observations 5:**
- The ones are the same variable
- ParentChildren and SiblingSpouse: have the highest positive number showing correlation
- Survived and Fare: the ones who paid the highest fare have the highest survival rate.
- Pclass and Survived: The Pclass goes up the lower survival rate.
- The 3rd class passengers had a lower survival rate.

```python
# Let's create the heatmap for this dataset
plt.figure(figsize = (12,8)) # Setting the size of the heatmap

sns.heatmap(num_df.corr(), annot=True, cmap = 'RdYlGn');
```
